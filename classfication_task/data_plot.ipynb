```python
# show dataset label distribution

import pandas as pd
from matplotlib import pyplot as plt

# 加载数据
data = pd.read_csv('/home/zhengyihao/dataset/covertype/covtype_with_key.subset.data')

print(data.shape[0])

# 使用 pandas 的 value_counts 函数计算每种 Cover_Type 的数量
cover_type_counts = data['Cover_Type'].value_counts()

# 画出条形图
cover_type_counts.plot(kind='bar')

# 设置标题和轴标签
plt.title('Distribution of Cover_Type')
plt.xlabel('Cover_Type')
plt.ylabel('Count')

# 显示图形
plt.savefig("/home/zhengyihao/TabularMark/classfication_task/plot/label_distribution")
```
