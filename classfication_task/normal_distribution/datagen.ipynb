```python
# generate mu = 0, sigma = 1 in normal distribution
import numpy as np
import pandas as pd

# 设定均值和标准差
mu, sigma = 0, 20 

# 生成数据
data = np.random.normal(mu, sigma, (2000, 2))

# 创建dataframe，并将列命名为dimension_0和dimension_1
df = pd.DataFrame(data, columns=['dimension_0', 'dimension_1'])

# 保存到csv文件
df.to_csv('synthetic_dataset.csv', index=False)

print("synthetic_dataset.csv generated!")
```

```python
# plot data distribution 
import matplotlib.pyplot as plt
import pandas as pd

# 读取数据
df = pd.read_csv('synthetic_dataset.csv')

# 创建子图
fig, axs = plt.subplots(2)

# 绘制dimension_0的分布
axs[0].hist(df['dimension_0'], bins=50, color='b', alpha=0.7)
axs[0].set_title('Distribution of dimension_0')

# 绘制dimension_1的分布
axs[1].hist(df['dimension_1'], bins=50, color='r', alpha=0.7)
axs[1].set_title('Distribution of dimension_1')

# 展示图像
plt.tight_layout()
plt.show()
```

```python
# generate labels
import numpy as np
import pandas as pd
from scipy.special import expit 

# 加载数据
df = pd.read_csv('synthetic_dataset.csv')
data = df.values

# 使用范围在[-1, 1]的随机权重
weights = -1 + 2*np.random.rand(2)

print(weights)

# 计算加权和
weighted_sum = np.dot(data, weights)

#获取事件发生的概率
prob = expit(weighted_sum)

# 生成标签
labels = np.random.binomial(1, prob)

# 将标签添加到dataframe
df['label'] = labels

# 保存更新后的数据集
df.to_csv('synthetic_dataset_labeled.csv', index=False)

print("labels generated!")
```

```python
# plot data distribution 
import pandas as pd
import matplotlib.pyplot as plt

# 读取数据
df = pd.read_csv('synthetic_dataset_labeled.csv')

# 获取标签数据
labels = df['label']

# 绘制直方图
plt.hist(labels, bins=[-0.5, 0.5, 1.5], edgecolor='black', rwidth=0.8)
plt.xticks([0, 1])
plt.title('Label Distribution')
plt.show()
```
